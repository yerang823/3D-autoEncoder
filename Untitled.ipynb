```python
import os, shutil
```

```python
path='../data/raw_32x32x32/'

li=os.listdir(path)
li
```

```python
cnt=0
for i in range(len(li)):
    x_size=int(li[i].split('.')[1].split('x')[0])
    if x_size <= 10:
        os.remove(path+li[i])
        print(path+li[i])
        cnt+=1
        
print(cnt)
```
